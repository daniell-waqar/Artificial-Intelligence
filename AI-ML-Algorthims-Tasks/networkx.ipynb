```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()    # directed graph
```

```python
G.add_edge(1,2)  # edge between one and two it also create two nodes 1 and 2
G.add_edge(2,3)
G.add_edge("A", "B")
G.add_edge("B", "B")
G.add_node("C")

```

```python
nx.draw_spring(G, with_labels=True)
```

```python
edge_list = [(1,2),(2,3),(3,4),(3,5),(4,6),(6,7)]
G = nx.from_edgelist(edge_list)
nx.draw_spring(G, with_labels=True)
```

```python
nx.draw_circular(G, with_labels=True)
```
